```python
# Survival Duration Calculator
def survival_duration_calculator(name):
  print (f"""
              ***Welcome {name} to the survival duration calculator***
              """)
  age = int(input("Enter your age "))
  if age == 0:
      print ("please enter a valid age thank you")
      return
  while True :

    print("""Choose the unit in which age to be calculate,
          1)month
          2)week
          3)days
          4)hours
          5)minutes
          6)seconds """)

    choice = int(input("Enter your choice in following unites: "))

    if choice == 1:
        age = age * 12
        print("You lived for", age, "months")

    elif choice == 2:
      age = age * 52
      print (" You lived for" , age , "week")

    elif choice == 3:
      age = age * 365
      print ( " You lived for",age, "days"  )

    elif choice == 4:
      age = age * 365 * 24
      print (" You lived for", age, "hours" )

    elif choice == 5:
      age = age * 365 * 24 * 60
      print ( " You lived for hours", age, "minutes" )

    elif choice == 6:
      age = age * 365 * 24 * 60 * 60
      print ( " You lived for", age, "seconds" )
    else:
      print("Enter a valid choice Exit")
      break
```

```python
name = "sandip"
```

```python
survival_duration_calculator(name)
```
